## 1. Implementación de Tracking de Métricas

#### 1.1 Sistema de Monitoreo Básico

```python
# Se prepara los datos (X = features numéricas, y = clases)
X = df_feat.select_dtypes(include=[np.number]).drop(
    columns=["class_code", "record", "sample"], errors="ignore"
)
y = df_feat["class_code"]

# División Train/Test estratificada
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=SEED
)

print("Train:", X_train.shape, " Test:", X_test.shape)
```

```python
# Modelo base: Regresión Logística
model = LogisticRegression(max_iter=1000, solver="saga", penalty="l2")

# Tracking de métricas con curva de aprendizaje (learning_curve)
train_sizes, train_scores, val_scores = learning_curve(
    model, X_train, y_train,
    cv=5, scoring="accuracy",
    train_sizes=np.linspace(0.1, 1.0, 5),
    n_jobs=-1
)

# Promedio y desviación
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)

# Gráfico: Accuracy en Train vs Validation según tamaño de datos
plt.figure(figsize=(8,6), dpi=300)
plt.plot(train_sizes, train_mean, "o-", color="teal", label="Train score")
plt.plot(train_sizes, val_mean, "o-", color="orange", label="Validation score")
plt.fill_between(train_sizes, train_mean-train_std, train_mean+train_std, alpha=0.2, color="teal")
plt.fill_between(train_sizes, val_mean-val_std, val_mean+val_std, alpha=0.2, color="orange")
plt.title("Curva de Aprendizaje (Logistic Regression)")
plt.xlabel("Tamaño del conjunto de entrenamiento")
plt.ylabel("Accuracy")
plt.legend(loc="best")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

#### Análisis de la Curva de Aprendizaje (Logistic Regression)

La figura muestra la evolución del **accuracy en entrenamiento y validación** a medida que aumenta el tamaño del conjunto de entrenamiento (10% → 100%).  

**Observaciones principales:**
- El rendimiento en **entrenamiento** comienza ligeramente superior (~0.805) pero desciende con el incremento de muestras, estabilizándose alrededor de **0.804–0.805**.  
- El rendimiento en **validación** se mantiene muy estable y con ligera tendencia ascendente conforme aumenta el tamaño de los datos, convergiendo con la curva de entrenamiento.  
- El **intervalo de confianza** (área sombreada) se reduce a medida que crece el tamaño de entrenamiento, lo que indica mayor estabilidad de la métrica con más datos.  

**Diagnóstico:**
- No se observa una brecha marcada entre entrenamiento y validación → **no hay señales claras de overfitting**.  
- Tampoco se percibe un accuracy bajo o sin mejora significativa → **no hay indicios de underfitting severo**.  
- Ambas curvas tienden a converger en un valor similar (~0.805), lo cual refleja un **ajuste razonable** para este modelo base.  

**Conclusión:**
El modelo de **Regresión Logística** logra un rendimiento estable y consistente en ambos conjuntos.  
La curva de validación sugiere que el modelo generaliza bien y que no se obtiene ganancia sustancial al aumentar más el tamaño de los datos en este escenario.

#### 1.2 Tracking para Diferentes Tipos de Modelos (Scikit-learn)

```python
# Se usa validation_curve para ver el efecto de un hiperparámetro (C)
param_range = np.logspace(-3, 2, 6)  # valores de C para regularización

# Modelo con semilla fija
model = LogisticRegression(max_iter=2000, solver="lbfgs", penalty="l2", random_state=SEED)

train_scores, val_scores = validation_curve(
    model, X_train, y_train,
    param_name="C", param_range=param_range,
    cv=5, scoring="accuracy", n_jobs=-1
)

# Promedio de scores
train_mean = np.mean(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)

# Gráfico: Accuracy en Train vs Validation según C
plt.figure(figsize=(8,6), dpi=300)
plt.semilogx(param_range, train_mean, "o-", color="teal", label="Train score")
plt.semilogx(param_range, val_mean, "o-", color="orange", label="Validation score")
plt.title("Validation Curve (Logistic Regression, parámetro C)")
plt.xlabel("Parámetro C (regularización inversa)")
plt.ylabel("Accuracy")
plt.legend(loc="best")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

#### Análisis de la Validation Curve (Logistic Regression, parámetro C)

La figura muestra cómo varía el **accuracy en entrenamiento y validación** al ajustar el hiperparámetro de regularización `C` en la Regresión Logística.

**Observaciones principales:**
- Con valores muy pequeños de `C` (alta regularización, `10⁻³`), tanto el entrenamiento como la validación presentan menor accuracy (~0.860).  
- A medida que `C` aumenta hacia `10⁻¹ – 10¹`, el rendimiento mejora, alcanzando un máximo de ~0.874 en validación.  
- Para valores de `C` demasiado grandes (`10²`), no se observa un beneficio adicional: las curvas se estabilizan e incluso muestran una ligera caída.  
- El **gap entre entrenamiento y validación es mínimo en todo el rango**, lo que indica buena generalización sin sobreajuste.

**Diagnóstico:**
- **Underfitting** en valores de `C` muy bajos (modelo demasiado restringido).  
- **Comportamiento óptimo** en el rango `C ≈ 0.1 – 10`, donde se logra el mejor equilibrio entre bias y varianza.  
- No hay evidencia de **overfitting severo**, ya que las curvas se mantienen cercanas y estables.

**Conclusión:**
El análisis sugiere que el hiperparámetro `C` debe elegirse en un rango intermedio (`10⁻¹` a `10¹`) para maximizar el rendimiento.  
Esto confirma que una regularización excesiva limita la capacidad del modelo, mientras que una regularización adecuada permite alcanzar el mejor balance entre ajuste y generalización.

## 2. Generación de Curvas de Aprendizaje

#### 2.1 Curvas Básicas Requeridas

```python
# A. Training vs Validation Loss

def plot_loss_curve(X_train, y_train, X_test, y_test, train_sizes=np.linspace(0.1, 1.0, 5)):
    """
    Genera la curva de pérdida (log-loss) para Logistic Regression.
    Incluye imputación de NaN y estandarización de variables.
    """
    # Modelo con imputación + escalado + regresión logística estable
    model = Pipeline([
        ("imputer", SimpleImputer(strategy="mean")),
        ("scaler", StandardScaler()),
        ("logreg", LogisticRegression(max_iter=2000, solver="lbfgs", penalty="l2"))
    ])
    
    train_losses, val_losses, sizes = [], [], []
    
    for frac in train_sizes:
        # Submuestreo del set de entrenamiento
        n_samples = int(frac * len(y_train))
        X_sub, y_sub = resample(X_train, y_train, replace=False, n_samples=n_samples, random_state=SEED)
        
        # Entrenamiento
        model.fit(X_sub, y_sub)
        
        # Predicciones probabilísticas
        y_train_pred = model.predict_proba(X_sub)
        y_val_pred = model.predict_proba(X_test)
        
        # Cálculo de log-loss
        train_losses.append(log_loss(y_sub, y_train_pred))
        val_losses.append(log_loss(y_test, y_val_pred))
        sizes.append(len(y_sub))
    
    # Gráfico
    plt.figure(figsize=(8,6), dpi=300)
    plt.plot(sizes, train_losses, "o-", color="teal", label="Entrenamiento (loss)")
    plt.plot(sizes, val_losses, "o-", color="orange", label="Validación (loss)")
    plt.title("Curva de Pérdida (Logistic Regression)")
    plt.xlabel("Tamaño del conjunto de entrenamiento")
    plt.ylabel("Log-Loss")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)
    
    # anotación del punto óptimo en validación (desplazada 20px arriba)
    best_idx = np.argmin(val_losses)
    plt.annotate(f"Mejor Validación: {val_losses[best_idx]:.2f}",
                 xy=(sizes[best_idx], val_losses[best_idx]),
                 xytext=(0, 20),              # offset en píxeles
                 textcoords="offset points",  # coordenadas relativas al punto
                 ha="center", color="black",
                 arrowprops=dict(arrowstyle="->", color="black"))
    
    plt.show()
    return train_losses, val_losses, sizes

# Ejecutar y capturar resultados
train_losses, val_losses, sizes = plot_loss_curve(X_train, y_train, X_test, y_test)

# Métricas de diagnóstico

# a) Accuracy promedio (de las curvas de aprendizaje)
print("Train Accuracy promedio:", train_mean.mean())
print("Validation Accuracy promedio:", val_mean.mean())
print("Gap promedio:", abs(train_mean.mean() - val_mean.mean()))

# b) Log-Loss mínimo
print("Min Validation Log-Loss:", min(val_losses))
print("Train Log-Loss final:", train_losses[-1])

# c) Overfitting ratio (gap relativo)
gap = abs(train_mean.mean() - val_mean.mean())
rel_gap = gap / val_mean.mean()
print("Relative Gap:", rel_gap)
```

#### Análisis de la Curva de Pérdida (Logistic Regression)

La figura muestra la evolución del **Log-Loss** en entrenamiento y validación a medida que aumenta el tamaño del conjunto de entrenamiento.

**Observaciones principales:**
- Tanto en entrenamiento como en validación, el **Log-Loss disminuye progresivamente** conforme aumenta el número de muestras, indicando una mejor capacidad de ajuste del modelo con más datos.
- El **Log-Loss de validación** es inicialmente superior al de entrenamiento (~ 0.436 vs ~ 0.432), pero ambas curvas convergen a valores cercanos (~ 0.426).
- La diferencia entre las curvas es muy pequeña y estable, sin evidencia de divergencia significativa.

**Métricas de diagnóstico:**
- **Accuracy promedio (train):** ~0.8712  
- **Accuracy promedio (val):** ~0.8715  
- **Gap promedio:** ~6.3e-05 (casi nulo).  
- **Min Validation Log-Loss:** ~0.426  
- **Train Log-Loss final:** ~0.4238  
- **Relative Gap:** ~7.2e-05  

**Diagnóstico:**
- No hay indicios de **overfitting**, ya que la validación sigue de cerca al entrenamiento.  
- Tampoco se observa **underfitting**, pues las pérdidas disminuyen y se estabilizan en valores bajos.  
- El comportamiento es consistente con un **modelo bien regularizado y en equilibrio**, con buena generalización.

**Conclusión:**
La curva de pérdida confirma que el modelo de **Regresión Logística** está en una zona de **ajuste óptimo**. El gap entre entrenamiento y validación es prácticamente nulo, lo que respalda que el modelo no sufre ni de sobreajuste ni de infraajuste significativo.

```python
# B. Training vs Validation Accuracy/Score

from sklearn.model_selection import learning_curve
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np

# Recalcular learning_curve para asegurar consistencia
train_sizes, train_scores, val_scores = learning_curve(
    LogisticRegression(max_iter=2000, solver="lbfgs", penalty="l2"),
    X_train, y_train,
    cv=5, scoring="accuracy",
    train_sizes=np.linspace(0.1, 1.0, 5),
    n_jobs=-1
)

# Promedios y desviaciones
train_mean, train_std = train_scores.mean(axis=1), train_scores.std(axis=1)
val_mean, val_std = val_scores.mean(axis=1), val_scores.std(axis=1)

# Plot
plt.figure(figsize=(8,6), dpi=300)
plt.plot(train_sizes, train_mean, "o-", color="teal", label="Entrenamiento (accuracy)")
plt.plot(train_sizes, val_mean, "o-", color="orange", label="Validación (accuracy)")
plt.fill_between(train_sizes, train_mean-train_std, train_mean+train_std, alpha=0.2, color="teal")
plt.fill_between(train_sizes, val_mean-val_std, val_mean+val_std, alpha=0.2, color="orange")
plt.title("Curva de Accuracy (Logistic Regression)")
plt.xlabel("Tamaño del conjunto de entrenamiento")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)

# Anotación en el mejor punto de validación (texto desplazado 20px arriba)
best_idx = np.argmax(val_mean)
plt.annotate(f"Mejor Validación: {val_mean[best_idx]:.2f}",
             xy=(train_sizes[best_idx], val_mean[best_idx]),
             xytext=(0, 20),  # 20 píxeles arriba
             textcoords="offset points",
             ha="center", color="black",
             arrowprops=dict(arrowstyle="->", color="black"))

plt.show()
```

#### Análisis de la Curva de Accuracy (Logistic Regression)

La figura muestra la evolución del **accuracy en entrenamiento y validación** a medida que aumenta el tamaño del conjunto de entrenamiento (10% → 100%).

**Observaciones principales:**
- El rendimiento en **entrenamiento** comienza ligeramente superior (~0.8745) y se mantiene estable en el rango **0.874–0.875**.
- El rendimiento en **validación** inicia algo más bajo (~0.873) pero muestra una **tendencia ascendente**, alcanzando un valor máximo cercano a **0.875** cuando se incrementa el tamaño de entrenamiento.
- Ambas curvas tienden a converger hacia un mismo valor, lo que refleja **consistencia y estabilidad** en el desempeño del modelo.
- El **intervalo de confianza** (área sombreada) se reduce con más datos, indicando que el modelo es más confiable a medida que crece el dataset.

**Diagnóstico:**
- No existe una brecha amplia ni creciente entre entrenamiento y validación → **no hay sobreajuste (overfitting)**.
- El accuracy es relativamente alto y mejora con más datos → **no hay infraajuste (underfitting)**.
- El comportamiento refleja un **ajuste óptimo**, con buena generalización en validación.

**Conclusión:**
La Regresión Logística logra un desempeño **sólido y balanceado** en ambos conjuntos. La curva de validación sugiere que el modelo aprovecha bien los datos disponibles, y que el rendimiento es estable y generaliza adecuadamente.

## 4. Estrategias de Mejora Propuestas

#### 4.1 Recomendaciones Específicas según Diagnóstico

Dado que el modelo base (Regresión Logística) no presenta señales claras de **overfitting** ni **underfitting**, las estrategias de mejora no se enfocan en corregir un problema crítico, sino en **explorar ajustes que puedan optimizar el rendimiento y robustez del modelo**:

- **Optimización de hiperparámetros de regularización (`C`, `penalty`)**  
  Ajustar `C` en el rango identificado como óptimo (10⁻¹ a 10¹) y comparar penalizaciones **L1** vs **L2** para buscar un mejor trade-off entre sparsity y estabilidad.

- **Feature Engineering y Selección de Variables**  
  Incorporar transformaciones adicionales (interacciones, variables derivadas) o aplicar selección de características (e.g., ANOVA F-test, modelos basados en importancia de características) para verificar si el modelo mejora al reducir ruido.

- **Validación cruzada más rigurosa (Stratified K-Fold con más splits)**  
  Si bien el modelo ya se validó, usar más folds aumenta la confianza estadística en el resultado y reduce la varianza de la estimación.

- **Comparación con modelos más complejos**  
  Usar el baseline de Regresión Logística como referencia y probar modelos más flexibles (Random Forest, XGBoost) para analizar si capturan mejor las relaciones no lineales.

---

#### 4.2 Implementación de Estrategias

Para cumplir con el requisito de implementar al menos dos estrategias, se proponen:

1. **Optimización de hiperparámetros con GridSearchCV:**  
   - Rango de `C = [0.01, 0.1, 1, 10]`  
   - Penalización: `['l1', 'l2']` (con `saga` o `liblinear` como solver).  
   - Métrica: F1-macro o accuracy.  

2. **Validación cruzada más robusta:**  
   - Uso de **StratifiedKFold con k=10** en lugar de k=5.  
   - Esto permite evaluar la estabilidad del modelo en más particiones y comprobar si el rendimiento se mantiene estable.

---

**Justificación:**  
Estas estrategias no atacan un “problema crítico”, pero sí permiten:
- Confirmar la robustez del modelo.  
- Detectar mejoras marginales en accuracy/log-loss.  
- Asegurar que el baseline está bien optimizado antes de compararlo con modelos más complejos.  



```python
# Optimización de hiperparámetros con GridSearchCV

# Pipeline con imputación + escalado + regresión logística
pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler()),
    ("logreg", LogisticRegression(max_iter=2000, solver="saga", random_state=SEED))
])

# Espacio de búsqueda
param_grid = {
    "logreg__C": [0.01, 0.1, 1, 10],
    "logreg__penalty": ["l1", "l2"]  # saga soporta L1 y L2
}

# GridSearchCV con validación cruzada estratificada
grid = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5,                     # 5 folds
    scoring="accuracy",       # métrica
    n_jobs=-1,                # usar todos los núcleos
    verbose=1
)

# Entrenamiento de la búsqueda
grid.fit(X_train, y_train)

# Resultados principales
print("=== Resultados GridSearchCV ===")
print("Mejores parámetros:", grid.best_params_)
print("Mejor accuracy en validación:", grid.best_score_)

# Visualización de resultados de GridSearchCV
results = pd.DataFrame(grid.cv_results_)

plt.figure(figsize=(8,6), dpi=300)
sns.pointplot(
    data=results,
    x="param_logreg__C", 
    y="mean_test_score", 
    hue="param_logreg__penalty", 
    dodge=True, markers="o", capsize=0.1
)
plt.xscale("log")
plt.title("Accuracy promedio por C y tipo de penalización (GridSearchCV)")
plt.xlabel("Parámetro C (escala log)")
plt.ylabel("Accuracy promedio (CV=5)")
plt.legend(title="Penalty")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

#### Análisis de Resultados con GridSearchCV (Regresión Logística)

La figura muestra el rendimiento promedio (accuracy en validación cruzada con 5 folds) de la **Regresión Logística** al variar el hiperparámetro de regularización `C` y el tipo de penalización (`L1` y `L2`).  

**Observaciones principales:**
- Existe una clara mejora en el rendimiento conforme aumenta el valor de `C` (menor regularización).  
- Para valores bajos de `C` (0.01 y 0.1), el accuracy promedio es menor (~0.871–0.873).  
- A partir de `C=1`, el rendimiento se estabiliza alrededor de **0.874**, sin ganancias sustanciales al incrementar más el valor de `C`.  
- Comparando penalizaciones:
  - **L1 (Lasso):** obtiene resultados competitivos, pero ligeramente inferiores o iguales a L2 en casi todos los valores de `C`.  
  - **L2 (Ridge):** ofrece el mejor resultado global con `C=1`, alcanzando un accuracy de **0.87417**.  

**Interpretación de L1 vs L2:**
- **L1 (Lasso):** penaliza la suma de los valores absolutos de los coeficientes, forzando a algunos a ser exactamente 0. Esto funciona como una forma de **selección automática de variables**, útil cuando existen muchas variables irrelevantes.  
- **L2 (Ridge):** penaliza el cuadrado de los coeficientes, haciendo que todos se reduzcan en magnitud pero rara vez lleguen a 0. Esto permite que **todas las variables aporten al modelo**, aunque con un peso ajustado para evitar sobreajuste.  

**Conclusión:**
El mejor modelo corresponde a **penalización L2 con `C=1`**, lo que confirma que probablemente **todas las features aportan información útil** y es más conveniente reducir sus pesos en conjunto en lugar de eliminar algunas por completo. Esto proporciona un modelo **más estable y balanceado**, adecuado para el dataset trabajado.

```python
# Validación cruzada más robusta (10 folds)
# Usando el mejor modelo encontrado en GridSearchCV
best_model = grid.best_estimator_

# Definir validación cruzada estratificada con 10 folds
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=SEED)

# Evaluar con cross_val_score
scores = cross_val_score(best_model, X_train, y_train, cv=cv, scoring="accuracy", n_jobs=-1)

# Resultados numéricos
print("\n=== Validación Cruzada Estratificada (10 folds) ===")
print("Accuracy promedio:", scores.mean())
print("Desviación estándar:", scores.std())

# Visualización de resultados de los folds
plt.figure(figsize=(6,6), dpi=300)
sns.boxplot(y=scores, color="teal")
sns.stripplot(y=scores, color="black", alpha=0.6)  # puntos individuales por fold
plt.title("Distribución de Accuracy en 10 folds (modelo optimizado)")
plt.ylabel("Accuracy")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

#### Validación Cruzada Estratificada con 10 folds (modelo optimizado)

Tras seleccionar los mejores hiperparámetros mediante GridSearchCV, se evaluó el modelo final utilizando una **validación cruzada estratificada con 10 folds**.  

**Resultados numéricos:**
- **Accuracy promedio:** 0.87437  
- **Desviación estándar:** 0.00223  

**Interpretación del gráfico (Boxplot):**
- El boxplot muestra la **distribución de los valores de accuracy** obtenidos en los 10 folds.  
- La mediana se ubica en torno a **0.874**, muy cercana al promedio global.  
- La dispersión es baja (σ ≈ 0.002), lo que indica una **alta estabilidad** del modelo frente a diferentes particiones de los datos.  
- No se observan valores atípicos extremos, lo que refuerza la consistencia de los resultados.  

**Conclusión:**
La validación cruzada con 10 folds confirma que el modelo optimizado es **robusto y generaliza adecuadamente**. La baja variabilidad entre folds indica que el desempeño no depende de una partición específica, lo cual da mayor confianza en su rendimiento en datos no vistos.
